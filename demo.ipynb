```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f_oneway, kruskal, test_ind, mannwhitneyu,normaltest

# Гама-розподіл
shape, scale = 0.3, 1.1

# моделювання зміни ціни акцій
def stock_price_at_time(t):
    price = 0  # Початкова ціна
    for _ in range(t):
        price += np.random.gamma(shape, scale)  # Використання гамма-розподілу
    return price

# запуск симуляції n разів
def simulate_n_times(n: int, t: int) -> list:
    np.random.seed(42)  # Фіксація випадковості для відтворюваності
    return [stock_price_at_time(t) for _ in range(n)]

# гістограми розподілу x
samples = np.random.gamma(shape, scale, 1000)
plt.hist(samples, bins=30, density=True, alpha=0.6, color='b', label="Гамма-розподіл")
plt.xlabel("Значення x")
plt.ylabel("Щільність ймовірності")
plt.title("Гістограма гамма-розподілу змінної x")
plt.legend()
plt.grid()
plt.show()

# симуляція для t від 1 до 60
t_values = list(range(1, 61, 2))  # Список значень t (крок 2)
data = {}  # Збереження результатів для аналізу

for t in t_values:
    results = simulate_n_times(100, t)
    mean_price = np.mean(results)
    k2, pvalue = normaltest(results) # додав !

    if pvalue > 0.05:
        print(f"✅ t={t}: Розподіл нормальний (p-value = {pvalue:.4f})")
```
